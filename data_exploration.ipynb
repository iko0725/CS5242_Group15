```python
from matplotlib import pyplot as plt
import numpy as np
import PIL
import cv2
```

```python
Fruits = ['apple', 'banana', 'strawberry', 'pear', 'watermelon', 'grapes','pineapple'] 
Flickr_num = []
Quickdraw_num = []
flickr_lst_width = []
quickdraw_lst_width = []
flickr_lst_height = []
quickdraw_lst_height = []
for fruit in Fruits:
  Flickr_count = 0
  Quickdraw_count = 0
  for i in range(10000):
    flickr_image_path = sys.path[0]+'/Flickr_scrape/'+ fruit
    quickdraw_image_path = sys.path[0]+'/quickdraw/'+ fruit
    flickr_image_path = os.path.join(flickr_image_path, str(i)+'.jpg')
    quickdraw_image_path = os.path.join(quickdraw_image_path, str(i)+'.jpg')
    if os.path.exists(flickr_image_path):
      Flickr_count += 1
      im = PIL.Image.open(flickr_image_path)
      width, height = im.size
      flickr_lst_width.append(width)
      flickr_lst_height.append(height)
    if os.path.exists(quickdraw_image_path):
      Quickdraw_count += 1
      im = PIL.Image.open(quickdraw_image_path)
      width, height = im.size
      quickdraw_lst_width.append(width)
      quickdraw_lst_height.append(height)
  Flickr_num.append(Flickr_count)
  Quickdraw_num.append(Quickdraw_count)
plt.hist([flickr_lst_width,quickdraw_lst_width], bins=25, label=['Flickr', 'Quickdraw'],)
plt.xlabel('Widths')
plt.ylabel('Number of images')
plt.legend(loc='upper right')
plt.tight_layout()
plt.title('Width distribution in the datasets')
plt.show()
plt.hist([flickr_lst_height,quickdraw_lst_height], bins=25, label=['Flickr', 'Quickdraw'],)
plt.xlabel('Heights')
plt.ylabel('Number of images')
plt.legend(loc='upper right')
plt.tight_layout()
plt.title('Height distribution in the datasets')
plt.show()
x = np.arange(len(Fruits))
bar_width  = 0.4
print(Flickr_num)
print(Quickdraw_num)
plt.bar(x, Flickr_num, bar_width, align="center", color='c', tick_label=Fruits, label='Flickr')
plt.bar(x+bar_width, Quickdraw_num, bar_width, align="center", color='y', tick_label=Fruits, label='Quickdraw')
plt.xlabel('classes')
plt.ylabel('Numbers per class')
plt.legend(loc='upper right')
plt.tight_layout()
plt.title('The number of images present in the datasets for each class')
```
